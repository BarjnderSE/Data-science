# What is Random Forest?

Random Forest is like a team of decision trees that work together to make a decision. Each tree gives its answer (vote), and the forest takes the majority vote (for classification) or average (for regression). This makes it more reliable and accurate than a single tree. It’s like asking 100 friends and going with the majority opinion!

Why is Random Forest Better?

    More Accurate: Combines many trees to reduce errors.
    Less Overfitting: Random Forest spreads the risk across trees.
    Reliable: Works better with noisy data (messy data).

If you're still confused, think of it as a cricket team:

    Decision Tree: One player batting for your team (might fail).
    Random Forest: Your whole team bats, and the average score is taken—more reliable!
    https://www.analyticsvidhya.com/blog/2021/06/understanding-random-forest/

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Extend classifier dictionary to include Random Forest
classifier = {
    
    "Random Forest Classifier": RandomForestClassifier(random_state=42)
}

# Loop through classifiers and evaluate
for name, clf in classifier.items():
    print(f"\n=========={name}===========")
    clf.fit(X_train, y_train)  # Train the model
    y_pred = clf.predict(X_test)  # Predict on test data
    
    # Evaluate metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    
    # Print metrics
    print(f"\nAccuracy: {accuracy:.2f}")
    print(f"Precision: {precision:.2f}")
    print(f"Recall: {recall:.2f}")
    print(f"F1 Score: {f1:.2f}")
```
